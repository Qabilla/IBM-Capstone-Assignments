# City Visualization and Analysis

Business/Problem Statement:
Using pincodes listed in websites to discover venues in a city (Mumbai) to determine the best location for opening a restaurant. The reason for
knowing other venues nearby is because to see if enough people are in the vicinity to become potential customers. The directory that will
be used will have a location information such as shopping malls, cafes, restaurants, etc. 

Data Processing:
Will be using pandas and foursquare to create datasets and package to generate geo locational data to review on where would be a suitable restaurant to be set up.
Note: pin codes and geo tags are read from India's pin code directory website.



Import all the nessary libraries.

```python
import pandas as pd
import numpy as np
import folium
from bs4 import BeautifulSoup
import requests
# !pip install pgeocode
# import pgeocode
from geopy.geocoders import Nominatim
```

Below code to extract data from the pin code website directory for Mumbai..

```python
html_doc = requests.get('https://finkode.com/mh/mumbai.html')
soup = BeautifulSoup(html_doc.content, 'html.parser')
pcode_table = soup.find('table',attrs={'class':'plist'})
pcode_table = pcode_table.find_all('tr')
del(pcode_table[0])
head_columns = ['Area','Pincode']
df = pd.DataFrame(columns=head_columns)
for tr in pcode_table:
  temp_data = tr.find_all('td')
  df = df.append({'Pincode':temp_data[2].text,'Area':temp_data[0].text}, ignore_index=True)
df.head()
```

```python
geolocator = Nominatim(user_agent="in_explorer")
for index,pincode in enumerate(df['Pincode']):
  code = geolocator.geocode('Mumbai - '+pincode)
  if code == None:
    lat = lon = 'NA'
  else:
    lat = code.latitude
    lon = code.longitude
  df.loc[index,'Latitude'] = lat
  df.loc[index,'Longitude'] = lon
  print(lat,lon)
df.head()
```

Remove NA..

```python
df = df[df['Latitude']!='NA']
df
```

Remove duplicates in table

```python
df.drop_duplicates(['Latitude','Longitude'],inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
address = 'Kalina, Mumbai'
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Mumbai are {}, {}.'.format(latitude, longitude))
```

```python
#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab

mum_map = folium.Map(location=[latitude,longitude],zoom_start=11)

for lat,lng,label in zip(df['Latitude'],df['Longitude'],df['Area']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
          [lat,lng],
          radius=5,
          popup=label,
          color='blue',
          fill=True,
          fill_color='#3186cc',
          fill_opacity=0.7,
          parse_html=False).add_to(mum_map)
mum_map
```
